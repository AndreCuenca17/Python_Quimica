```python
df = pd.read_csv("data.csv")
df
```

```python

df = df.rename(columns={'x': 'Volumen', 'y': 'Presion'})
df
```

```python
plt.scatter(df["Volumen"],df["Presion"])
plt.ylabel("Presion")
plt.xlabel("V")
```

```python
!pip install scipy
```

```python
from scipy import stats
```

```python
x = df["Volumen"]
y = df["Presion"]

pendiente, intercepto, r_value, p_value, std_err = stats.linregress(x, y)

print(f"Ecuación de la línea de regresión: y = {pendiente:.2f}x + {intercepto:.2f}")
```

```python
plt.scatter(x, y)
plt.plot(x, pendiente * x + intercepto, label="Línea de Regresión", color="red")
plt.ylabel("Presión")
plt.xlabel("Volumen")
plt.legend()
plt.show()
```

```python
df2 = pd.read_csv("data (2).csv")
#el rename no cambia el data frame directamente sino que devuelve uno nuevo
df2 = df2.rename(columns={'x':'T', 'y':'V'})
df2
```

```python
x2 = df2["T"]
y2 = df2["V"]

slope, intercept, r_value, p_value, std_err = stats.linregress(x2, y2)
print(f"Ecuación de la línea de regresión: y = {slope:.2f}x + {intercept:.2f}")
```

```python
plt.scatter(x2, y2)
plt.plot(x2, slope * x2 + intercept, label="Línea de Regresión", color="red")
plt.ylabel("Presión(ATM)")
plt.xlabel("Temperatura(Celsius)")
plt.legend()
plt.show()
```
